```python
# nice_house = cv2.imread("D:/git-repos/pylogoburner/imgs/nice-house.png")
nice_house = cv2.imread("D:/amigos/maria_de_fatima/house_imgs/WhatsApp Image 2022-12-07 at 18.04.17.jpeg")
plt.imshow(cv2.cvtColor(nice_house, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
h_img, w_img, _ = nice_house.shape
```

```python
scale_percent = 60 # percent of original size
width = int(logo.shape[1] * scale_percent / 100)
height = int(logo.shape[0] * scale_percent / 100)
dim = (width, height)
  
# resize image
logo = cv2.resize(logo, dim, interpolation = cv2.INTER_AREA)
```

```python
h_logo, w_logo, _ = logo.shape
logo.shape,nice_house.shape
```

```python
center_y = int(h_img / 2)
center_x = int(w_img / 2)
center_y,center_x
```

```python
top_y = center_y - int(h_logo / 2)
bottom_y = top_y + h_logo
left_x = center_x - int(w_logo / 2)
right_x = left_x + w_logo

(top_y,bottom_y,left_x,right_x)
```

```python
# Get ROI
roi = nice_house[top_y:bottom_y, left_x:right_x]
# nice_house[top_y:bottom_y, left_x:right_x] = logo
```

```python
result = cv2.addWeighted(roi, 1, logo, 0.2, 0)
```

```python
nice_house[top_y:bottom_y, left_x:right_x] = result
```

```python
plt.imshow(cv2.cvtColor(nice_house, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Save the modified image
cv2.imwrite("D:/amigos/maria_de_fatima/house_imgs_logoburn/my_first_image.jpg", nice_house)
```
